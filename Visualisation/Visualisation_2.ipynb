```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import seaborn as sns
```

# Avis positif vs negatifs

```python
df = pd.read_csv('../data.csv', sep=';', index_col='Unnamed: 0')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# perday plot nb avis positif, negatif total
Avis_Par_Jour = pd.DataFrame(df.groupby(['Date','Avis'], sort='Day')['numGS'].sum()).reset_index() 

pos = list(Avis_Par_Jour[Avis_Par_Jour['Avis'] == 'pos']['numGS'])
neg = list(-Avis_Par_Jour[Avis_Par_Jour['Avis'] == 'neg']['numGS'])
```

```python
# affichage des jours:
# via matplotlib 

x = list(Avis_Par_Jour['Date'].apply(mdates.strpdate2num('%Y-%m-%d')).unique())


y = np.row_stack((pos, neg))


fig, ax = plt.subplots(figsize=(15,10))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(x, rotation='vertical')

ax.stackplot(x, y, baseline='zero', labels = ('pos', 'neg'))
#on ajoute une ligne a la date des elections:
ax.axvline(dt.datetime(2017, 5, 7), c='red')
ax.axvline(dt.datetime(2017, 4, 23), c='red')
ax.legend()
plt.show()
```

# generalisation par pays:




```python
df = pd.read_csv('../data.csv', sep=';', index_col='Unnamed: 0')
```

```python
Avis_Par_Jour_Par_Pays = pd.DataFrame(df.groupby(['Date','Country_Code','Avis'], sort=['Day','Country_Code'])
                                      ['numGS'].sum()).reset_index() 

```

```python
dpos = Avis_Par_Jour_Par_Pays[Avis_Par_Jour_Par_Pays['Avis'] == 'pos'].pivot(index='Country_Code', columns='Date', values='numGS')
dneg = Avis_Par_Jour_Par_Pays[Avis_Par_Jour_Par_Pays['Avis'] == 'neg'].pivot(index='Country_Code', columns='Date', values='numGS')
```

```python
dpos_sum = np.sum(dpos,axis=1)
dpos_sum_sorted = dpos_sum.sort_values(axis=0, ascending=False)
top_10 = pd.DataFrame(dpos_sum_sorted.head(11))
top_10_pos = set(top_10.index)
top_10_pos
dneg_sum = np.sum(dneg,axis=1)
dneg_sum_sorted = dneg_sum.sort_values(axis=0, ascending=False)
top_10 = pd.DataFrame(dneg_sum_sorted.head(11))
top_10_neg = set(top_10.index)
top_10_neg
top = top_10_pos.intersection(top_10_neg)
liste_Pays = list(top)
liste_Pays
```

```python
# pour chaque type d'avis on construit un data frame pivot pour avoir un pay par ligne, 
# et en colonne, les valeur NumGS
dpos = Avis_Par_Jour_Par_Pays[(Avis_Par_Jour_Par_Pays['Avis'] == 'pos') & (Avis_Par_Jour_Par_Pays['Country_Code'].isin(liste_Pays))].pivot(index='Country_Code', columns='Date', values='numGS')
dneg = Avis_Par_Jour_Par_Pays[(Avis_Par_Jour_Par_Pays['Avis'] == 'neg') & (Avis_Par_Jour_Par_Pays['Country_Code'].isin(liste_Pays))].pivot(index='Country_Code', columns='Date', values='numGS')

#on rempli les NA par 0
dpos = dpos.fillna(0)
dneg = dneg.fillna(0)

#on complete pour avoir tous les pays dans les 2 matrices en index, le groupby peut faire perdre des pays:
dpos = dpos.append(pd.DataFrame(0, index= (set(liste_Pays) - set(dpos.index)), columns=dpos.columns))
dneg = dneg.append(pd.DataFrame(0, index= (set(liste_Pays) - set(dneg.index)), columns=dneg.columns))

#on ordonne les matrices par pays (index)
dpos.sort_index(inplace=True)
dneg.sort_index(inplace=True)
liste_Pays = np.sort(liste_Pays)
```

```python
#sharing same axe

colors = ['g', 'w', 'r', 'c', 'm', 'y', 'k', 'b']
x = list(Avis_Par_Jour_Par_Pays['Date'].apply(mdates.strpdate2num('%Y-%m-%d')).unique())
ypos = dpos
yneg = dneg



plt.close('all')


# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True, figsize=(15,10))
plt.stackplot([],[],colors='b',labels='USA')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(x, rotation='vertical')

axarr[0].stackplot(x, ypos, colors=colors, labels=liste_Pays)
axarr[0].set_title("nombre d'avis positif")
axarr[1].stackplot(x, -yneg, colors=colors)
axarr[1].set_title("nombre d'avis negatif")

#on ajoute une ligne a la date des elections:
axarr[0].axvline(dt.datetime(2017, 5, 7), c='red')
axarr[1].axvline(dt.datetime(2017, 5, 7), c='red')
axarr[0].axvline(dt.datetime(2017, 4, 23), c='red')
axarr[1].axvline(dt.datetime(2017, 4, 23), c='red')
f.subplots_adjust(hspace=0.3)
axarr[0].legend()
plt.show()
```
